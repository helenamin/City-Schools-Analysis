```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head(10)
```

## Player Count

* Display the total number of players


```python
player_data_df = purchase_data
player_data_df = player_data_df[["SN","Gender","Age"]].drop_duplicates()
player_dicts=[{"Total Players":len(player_data_df)}]
playerCount_df= pd.DataFrame(player_dicts)
playerCount_df
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
uniqueItems = purchase_data["Item Name"].unique()

avgPrice = purchase_data["Price"].mean()

purchaseCount = len(purchase_data["Purchase ID"])

totalRevenue = purchase_data["Price"].sum()


PurchasingAnalysis = [{"Number of Unique Items":len(uniqueItems),
                      "Average Price":avgPrice,
                      "Number of Purchases":purchaseCount,
                      "Total Revenue":totalRevenue}]
PurchasingAnalysis_df = pd.DataFrame(PurchasingAnalysis)
format_dict = {"Average Price":"${:.2f}","Total Revenue":"${:.2f}"}
PurchasingAnalysis_df.style.format(format_dict)
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
Gender_df= purchase_data.groupby(["Gender"])

Gender_Demographics = Gender_df["SN"].nunique()

Gender_Demographics_df = pd.DataFrame(Gender_Demographics)


Gender_Demographics_df= Gender_Demographics_df.rename(columns={"SN":"Total Count"})

Gender_Demographics_df["Percentage of Players"] = (Gender_Demographics_df["Total Count"])/(Gender_Demographics_df["Total Count"].sum())

Gender_Demographics_df["Percentage of Players"] = Gender_Demographics_df["Percentage of Players"].map("{:.2%}".format)
Gender_Demographics_df.sort_values(["Total Count"],ascending = False)

```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python

purchase_count_byGender = Gender_df["Price"].count()
purchase_count_byGender

purchase_avg_byGender = Gender_df["Price"].mean()
purchase_avg_byGender

purchase_total_byGender = Gender_df["Price"].sum()
purchase_total_byGender


Purchasing_Analysis_byGender_df = pd.DataFrame(purchase_count_byGender)
Purchasing_Analysis_byGender_df["Average Purchase Price"] = Gender_df["Price"].mean()
Purchasing_Analysis_byGender_df["purchase_total_byGender"] = Gender_df["Price"].sum()
Purchasing_Analysis_byGender_df["Avg Total Purchase per Person"] = Purchasing_Analysis_byGender_df["purchase_total_byGender"]/Gender_Demographics_df["Total Count"]

format_dict_2 = {"Average Purchase Price":"${:.2f}","purchase_total_byGender":"${:.2f}","Avg Total Purchase per Person":"${:.2f}"}
Purchasing_Analysis_byGender_df.style.format(format_dict_2)
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

age_data_df = player_data_df.loc[:,["SN","Age"]]

age_data_df["Age Category"] = pd.cut(age_data_df["Age"],bins,labels=group_labels,include_lowest = True)


age_df = age_data_df.groupby("Age Category").nunique()
del age_df["Age"]

age_df = age_df.rename(columns={"SN": "Total Counts"})

age_df["Percentage of Players"] = (age_df["Total Counts"]/(age_df["Total Counts"].sum()))

format_dict_1 = {"Percentage of Players":"{:.2%}"}
age_df.style.format(format_dict_1)
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
bins = [0,9.9,14.9,19.9,24.9,29.9,34.9,39.9,100]
group_labels = ["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

purchase_data_byAge = purchase_data
purchase_data_byAge["Age Category"] = pd.cut(purchase_data_byAge["Age"],bins,labels=group_labels,include_lowest = True)

purchase_Analysis_byAge = pd.DataFrame(purchase_data_byAge.groupby("Age Category")["Purchase ID"].nunique())
purchase_Analysis_byAge["Average Purchase Price"]= purchase_data_byAge.groupby("Age Category")["Price"].mean()
purchase_Analysis_byAge["Total Purchase Value"]= purchase_data_byAge.groupby("Age Category")["Price"].sum()
purchase_Analysis_byAge["Avg Total Purchase per Person"]= purchase_data_byAge.groupby("Age Category")["Price"].sum()/age_df["Total Counts"]

format_dict_3 = {"Average Purchase Price":"${:.2f}","Total Purchase Value":"${:.2f}","Avg Total Purchase per Person":"${:.2f}"}
purchase_Analysis_byAge.style.format(format_dict_3)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

spenders_df  = pd.DataFrame(purchase_data.groupby("SN")["Purchase ID"].nunique())
spenders_df = spenders_df.rename(columns={"Purchase ID":"Purchase Count"})
spenders_df["Average Purchase Price"] = purchase_data.groupby("SN")["Price"].mean()
spenders_df["Total Purchase Value"] = purchase_data.groupby("SN")["Price"].sum()

spenders_df.sort_values(["Total Purchase Value"],ascending = False,inplace=True)

spenders_df["Average Purchase Price"]=spenders_df["Average Purchase Price"].map("${:.2f}".format)
spenders_df["Total Purchase Value"] = spenders_df["Total Purchase Value"].map("${:.2f}".format)

spenders_df
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
df = pd.DataFrame(purchase_data.groupby(["Item ID","Item Name"]).agg({"Item Name":"count","Price":"mean"}))
df["Total Purchase Value"] = purchase_data.groupby(["Item ID","Item Name"])["Price"].sum()
df = df.rename(columns={"Item Name":"Purchase Count","Price":"Item Price"})

MostPopularItems_df= df.sort_values(["Purchase Count"],ascending = False)

MostPopularItems_df["Item Price"] = MostPopularItems_df["Item Price"].map("${:.2f}".format)
MostPopularItems_df["Total Purchase Value"] = MostPopularItems_df["Total Purchase Value"].map("${:.2f}".format)


MostPopularItems_df.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
MostProfitableItems_df = df.sort_values(["Total Purchase Value"],ascending = False)

MostProfitableItems_df["Item Price"] = MostProfitableItems_df["Item Price"].map("${:.2f}".format)
MostProfitableItems_df["Total Purchase Value"] = MostProfitableItems_df["Total Purchase Value"].map("${:.2f}".format)


MostProfitableItems_df.head()
```
